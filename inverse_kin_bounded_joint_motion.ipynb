```python
# Import Necessary Packages
from ikpy.chain import Chain
from ikpy.link import OriginLink, URDFLink
import matplotlib.pyplot
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
```

```python
# Manually Create Joints
FL_chain = Chain(name='front_left', links=[
    OriginLink(),
    URDFLink(
      name="hip",
      origin_translation=[0.1881, 0.04675, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="thigh",
      origin_translation=[0, 0.08, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="calf",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="foot",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    )
])
FR_chain = Chain(name='front_right', links=[
    OriginLink(),
    URDFLink(
      name="hip",
      origin_translation=[0.1881, -0.04675, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
      bounds=[-0.802851455917, 0.802851455917],
    ),
    URDFLink(
      name="thigh",
      origin_translation=[0, -0.08, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
        bounds=[-1.0471975512, 4.18879020479],
    ),
    URDFLink(
      name="calf",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
        # CODE WORKS WHEN THE STRICT BOUND IS REMOVED
        #bounds=[-7, 7], 
        bounds=[-2.69653369433, -0.916297857297],
    ),
    URDFLink(
      name="foot",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    )
])
RL_chain = Chain(name='rear_left', links=[
    OriginLink(),
    URDFLink(
      name="hip",
      origin_translation=[-0.1881, 0.04675, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
        
    ),
    URDFLink(
      name="thigh",
      origin_translation=[0, 0.08, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="calf",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="foot",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    )
])
RR_chain = Chain(name='rear_right', links=[
    OriginLink(),
    URDFLink(
      name="hip",
      origin_translation=[-0.1881, -0.04675, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="thigh",
      origin_translation=[0, -0.08, 0],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="calf",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    ),
    URDFLink(
      name="foot",
      origin_translation=[0, 0, -0.213],
      origin_orientation=[0, 0, 0],
      rotation=[0, 1, 0],
    )
])
```

```python
# Plot Initial Chains
ax = matplotlib.pyplot.figure().add_subplot(111, projection='3d')
#ax.set_color("black")
#left_arm_chain.plot(left_arm_chain.inverse_kinematics([2, 2, 2]), ax)
#ax2 = matplotlib.pyplot.figure().add_subplot(111, projection='3d', color = (0.1, 0.9, 0.5))
#left_arm_chain.plot(left_arm_chain.inverse_kinematics([1, -1, 4]), ax)
#ax3 = matplotlib.pyplot.figure().add_subplot(111, projection='3d', color = (0.9, 0.2, 0.5))
FR_chain.plot([0, 0, 0, -1, 0], ax)
FL_chain.plot([0] * (len(FL_chain)), ax)
RL_chain.plot([0] * (len(RL_chain)), ax)
RR_chain.plot([0] * (len(RR_chain)), ax)
matplotlib.pyplot.show()
```
